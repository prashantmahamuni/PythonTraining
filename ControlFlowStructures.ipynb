## If Else

    1.If statement is used to decide whether a certain statement should be executed or not
    2.Else statement will be executed if 'if-statement' condition is false
   


```python
x = 17

if x!=1:
    print('One')
else:
    print('x is something else')
```

## Nested If-Else

```python
x = 4

if x==1:
    print('One')
elif x==2:
    print('Two')
elif x==3:
    print('Three')
else:
    print('X is neither one, two or three')
    print('2nd print')
```

## Looping

For Loop:

    1.Used for traversing sequence
    2.Iterates over iterable
    
While Loop:

    1.Excutes block of statement until condition is true
    2.Exit condition needs to be defined carefully otherwise loop will go infinite and go OOM

```python
# For Loop
lst = [1, 2, 3, 4]

for i in lst:
    print('Value of i is =',i)
```

```python
for i in range(1, 100, 5):
    print(i)
```

```python
# While Loop
i = 1

while i < 10:
    print('Value of i is =',i)
    i += 1 # i = i + 1
```

```python
while True:
    print('Hi')
    
    break
```

## Break, Continue and Pass

Break:

    1.Break the loop 
    2.Break will end current loop and exit from the loop
    
Continue:

    1.Continue will exit current iteration
    2.Loop will continue for next iteration
    3.Wont end loop
    
Pass:
    
    1.Used to denote no code for execution

```python
# Break

stock = 5
x = int(input("How many candy's do you want?")) #10

i=1
while i <= x:
    if i > stock:
        print('Out of Stock')
        break
    
    print('Candy:',i)
    i += 1 # 2, 3, 4, 5, 
```

```python
# Continue
# print 1 to 20 values but skip divisible by 3 

for i in range(1, 21):
    if i % 3 == 0:
        continue
    elif i % 8 ==0:
        break
    else:
        print(i)
```

```python
# Pass
for i in range(1, 21):
    pass
```
