## Try Except Block to catch and handle exception

```python
# Catching very specific exception
a = 10
try:
    print(a/0)
except ZeroDivisionError:
    print("Please do not attempt to divide number by zero")
```

```python
# It is not a good practice but if you are not sure about exception type which may encouter the please feel free to use broader exctption category
# e.g. let's access try to access item from directory with incorrect key

dir1 = {'Name' : "John", 'age' : 30, 'Dept' : 'IT'}

Salary = dir1['Salary']
```

```python
# Handling any exception with broader exception category


dir1 = {'Name' : "John", 'age' : 30, 'Dept' : 'IT'}

try:
    Salary = dir1['Salary']
except Exception:
    print('Error occured')
    print(Exception.with_traceback()) ## How to write error message?
```

## Finally

+ The finally block will be executed no matter if the try block raises an error or not. 
+ This can be useful to close objects and clean up resources.

```python
try:
    f = open('./input/username.csv')
    content = f.read()
    print(content)
except Exception:
    print('Some exception happened')
finally:
    try:
        f.close()
    except:
        print('File has not opened')

        
```

## User defined or Custom exception

```python
# define Python user-defined exceptions

class InvalidAgeException(Exception):
    
    def __str__(self):
        return "Age must be greater than 18 for Voting"

```

```python
ageLimit = 18

input_num = int(input("Enter a age: "))
if input_num < ageLimit:
    raise InvalidAgeException
else:
    print("Eligible to Vote")
        
```

```python
ageLimit = 18

try:
    input_num = int(input("Enter a number: "))
    if input_num < ageLimit:
        raise InvalidAgeException
    else:
        print("Eligible to Vote")
        
except InvalidAgeException:
    print("Exception occurred: Invalid Age")
```

```python
def checkAge(input_num):
    try:
        if input_num < ageLimit:
            raise InvalidAgeException
        else:
            print("Eligible to Vote")        
    except InvalidAgeException:
        print("Exception occurred: Invalid Age")
        
    
```

```python
for userAge in [17, 18, 19, 20]:
    checkAge(userAge)
```
