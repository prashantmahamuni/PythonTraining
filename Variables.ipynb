Variables are containers or placeholders to store the data value.


Datatypes in Python

- None
- Numeric
    - integer
    - float
    - complex
    - boolean
- Sequence
    - List
    - Tuple
    - Set
    - String
    - Range
- Dictionalry

```python
# Variable of type None
abc = None
print(abc)
print(type(abc))
```

```python
# numeric variables - int
x = 10
print(x)
print(type(x))
```

```python
# numeric variables - float
y = 10.19
print(y)
print(type(y))
```

```python
# numeric variables - boolean
z = True
print(z)
print(type(z))
```

```python
# numeric variables - complex
type(10 + 20j)   #j is imaginary number
```

```python
# Sequence variables - list
list1 = [1, 2, 'a', True]
print(list1)
print(type(list1))
```

```python
# Sequence variables - tuple
tuple1 = (1, 2, 'a', True)
print(tuple1)
print(type(tuple1))
```

```python
# Sequence variables - set
set1 = {1, 2, 'a'}
print(set1)
print(type(set1))
```

```python
# Sequence variables' - String
String1 = 'Hello'
print(String1)
print(type(String1))
```

```python
# Sequence variables' - range
range1 = range(1,10)
print(range1)
print(type(range1))
```

```python
# Dictionary

dict1 = {"color": "red", "wheels": "four", "brand": "Toyota", "model":"2019"}
print(dict1)
print(type(dict1))
```
